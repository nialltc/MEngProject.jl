```julia
mutable struct LamFunction_mtk{T <: AbstractArray} <: Function
	x::T
	m::T
	s::T
	
	x_lgn::T
	C::T
	H_z::T
	
	dy_temp::T
	dm_temp::T
	dz_temp::T
	ds_temp::T
	dv_temp::T
	H_z_temp::T
	V_temp_1::T
	V_temp_2::T
	Q_temp::T
	P_temp::T
    pp::NamedTuple
end


function (ff::LamFunction_mtk)(du, u, p, t)

       @inbounds begin

		@. ff.x = @view u[:, :, 1:ff.pp.K,:]
        y = @view u[:, :, ff.pp.K+1:2*ff.pp.K,:]
        @. ff.m = @view u[:, :, 2*ff.pp.K+1:3*ff.pp.K,:]
        z = @view u[:, :, 3*ff.pp.K+1:4*ff.pp.K,:]
		@. ff.s = @view u[:, :, 4*ff.pp.K+1:5*ff.pp.K,:]

		v_p = @view u[:, :, 5*ff.pp.K+1:5*ff.pp.K+1,:]
		v_m = @view u[:, :, 5*ff.pp.K+2:5*ff.pp.K+2,:]
		
		dx = @view du[:, :, 1:ff.pp.K,:]
        dy = @view du[:, :, ff.pp.K+1:2*ff.pp.K,:]
        dm = @view du[:, :, 2*ff.pp.K+1:3*ff.pp.K,:]
        dz = @view du[:, :, 3*ff.pp.K+1:4*ff.pp.K,:]
        ds = @view du[:, :, 4*ff.pp.K+1:5*ff.pp.K,:]

		dv_p = @view du[:, :, 5*ff.pp.K+1:5*ff.pp.K+1,:]
		dv_m = @view du[:, :, 5*ff.pp.K+2:5*ff.pp.K+2,:]
		
		LaminartGPU.fun_x_lgn!(ff.x_lgn, ff.x, ff.pp)
        LaminartGPU.fun_v_C!(ff.C, v_p, v_m, ff.V_temp_1, ff.V_temp_2, ff.Q_temp, ff.P_temp, ff.pp)
        LaminartGPU.fun_H_z!(ff.H_z, z, ff.H_z_temp, ff.pp)

        LaminartGPU.fun_dv!(dv_p, v_p, ff.pp.r, ff.x_lgn, ff.dv_temp, ff.pp)
       LaminartGPU.fun_dv!(dv_m, v_m, .-ff.pp.r, ff.x_lgn, ff.dv_temp, ff.pp)
        LaminartGPU.fun_dx_v1!(dx, ff.x, ff.C, z, ff.pp.x_V2, ff.pp)
        LaminartGPU.fun_dy!(dy, y, ff.C, ff.x, ff.m, ff.dy_temp, ff.pp)
        LaminartGPU.fun_dm!(dm, ff.m, ff.x, ff.dm_temp, ff.pp)
        LaminartGPU.fun_dz!(dz, z, y, ff.H_z, ff.s, ff.dz_temp, ff.pp)
        LaminartGPU.fun_ds!(ds, ff.s, ff.H_z, ff.ds_temp, ff.pp)
		
    end
    return nothing

end
```
